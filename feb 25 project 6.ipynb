```python
 import scipy.stats
```

```python
x = df.Population
y = df.Literate
z = df.Christians
scipy.stats.spearmanr(x, y)
xyz = np.array([df['Population'], df['Literate'], df['Christians'], df['Secondary_Education']])
```

```python
corr_matrix, p_matrix = scipy.stats.spearmanr(xyz, axis=1)
corr_matrix
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
corr_matrix = np.corrcoef(xyz).round(decimals=2)
corr_matrix
```

```python
fig, ax = plt.subplots()
im = ax.imshow(corr_matrix)
im.set_clim(-1, 1)
ax.grid(False)
ax.set_ylim(2.5, -0.5)
for i in range(4):
    for j in range(4):
        ax.text(j, i, corr_matrix[i, j], ha='center', va='center',
                color='r')
        
cbar = ax.figure.colorbar(im, ax=ax, format='% .2f')
plt.show()
```

```python

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] =  df_numerized[col_name].astype('category')
        df_numerized[col_name] =  df_numerized[col_name].cat.codes
df_numerized  
```

```python

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,20)) 
correlation_matrix = df_numerized.corr(method = 'spearman')
sns.heatmap(correlation_matrix, annot= True, linewidths=2)
plt.show()
```
